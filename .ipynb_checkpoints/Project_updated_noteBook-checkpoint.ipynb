#### 1. Motivation
The motivation behind this project was to explore the relationship between various socio-economic factors and overall happiness scores across different countries. Given the complexity of well-being, understanding these relationships can provide insights into policy making and personal priorities.

#### What is your Dataset?
The dataset, named "Updated_Merged_Happiness_Report_2015_2023.csv," appears to be a compilation of annual happiness scores along with socio-economic indicators like GDP per capita, social support, and health metrics across multiple countries over several years.

### Handeling missing values

```python
print("Missing values:")
print(df.isna().sum())

#Handelling missing values
# Filling the missing value with the mean
df.fillna(df['healthy_life_expectancy'].mean(),inplace=True)

print("--------------------------------------------")
print("After handeling with the missing values")
print(df.isnull().sum())
```

#### 2.  Basic stats
##### Data Cleaning and Preprocessing
Resolving missing values, standardizing variable formats, and guaranteeing consistency across several years were all part of the data-cleaning procedure for longitudinal research. Approaches have been implemented to reduce biases and inaccuracies while preserving transparency and promoting trust in the results of longitudinal studies.

### GDP and Happiness

```python
numeric_data = df.select_dtypes(include='number')
correlation_with_happiness = numeric_data.corrwith(numeric_data['happiness_score'])
print("Correlation with happiness")
print("\n" + "------------------------------------" )
print(correlation_with_happiness)
print("------------------------------------")
```

```python
correlation = df.corr()
import seaborn as sns
sns.heatmap(correlation, annot = True, cmap="plasma")
```

```python
fig, ax = plt.subplots(4,2, figsize=(14, 14))  

sns.histplot(df['happiness_score'], bins= 40, kde= True, ax=ax[0,0])
sns.histplot(df['gdp_per_capita'], bins= 40, kde= True, ax=ax[0,1])
sns.histplot(df['social_support'], bins= 40, kde= True, ax=ax[1,0])
sns.histplot(df['healthy_life_expectancy'], bins= 40, kde= True, ax=ax[1,1])
sns.histplot(df['freedom_to_make_life_choices'], bins= 40, kde= True, ax=ax[2,0])
sns.histplot(df['generosity'], bins= 40, kde= True, ax=ax[2,1])
sns.histplot(df['perceptions_of_corruption'], bins= 40, kde= True, ax=ax[3,0])

fig.suptitle('Data distribution')
ax[0,0].set_xlabel('appiness_score')
ax[0,1].set_xlabel('gdp_per_capita')
ax[1,0].set_xlabel('social_support')
ax[1,1].set_xlabel('healthy_life_expectancy')
ax[2,0].set_xlabel('freedom_to_make_life_choices')
ax[2,1].set_xlabel('generosity')
ax[3,0].set_xlabel('perceptions_of_corruption')
plt.show()
```

```python
fig, ax = plt.subplots(2,3, figsize= (14, 10))
fig.suptitle('Linear dependancy with happiness_score')

sns.regplot(data=df, x='happiness_score', y='gdp_per_capita', ax=ax[0,0])
sns.regplot(data=df, x='happiness_score', y='social_support', ax=ax[0,1])
sns.regplot(data=df, x='happiness_score', y='healthy_life_expectancy', ax=ax[0,2])
sns.regplot(data=df, x='happiness_score', y='freedom_to_make_life_choices', ax=ax[1,0])
sns.regplot(data=df, x='happiness_score', y='generosity', ax=ax[1,1])
sns.regplot(data=df, x='happiness_score', y='perceptions_of_corruption', ax=ax[1,2])

plt.show()
```

```python
# Calculate mean happiness scores and mean GDP per capita by country, sort by happiness scores in descending order, and reset the index
sorted_happiness_scores = df.groupby('country').agg({
    'happiness_score': 'mean',
    'gdp_per_capita': 'mean'
}).sort_values(by='happiness_score', ascending=False).reset_index()

# Rename the columns to more descriptive names
sorted_happiness_scores.columns = ['country', 'average_happiness_score', 'average_gdp_per_capita']
sorted_happiness_scores
```
